## Hypothesis


We predict study time will be the most significant predictor of both GPA and stress levels, while sleep time will likely be positively correlated with GPA and negatively correlated with stress levels and social time will be negatively correlated with both. We expect higher exercise levels and extracurricular time to be predictive of lower stress levels, but not significantly correlated with GPA.
